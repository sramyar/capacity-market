```python
# importing packages

import numpy as np
import pandas as pd
from scipy.linalg import cholesky
from scipy.stats import norm
from pylab import plot, show, axis, subplot, xlabel, ylabel, grid

# set sample size
sample_size = 200

# set covariance matrix

cov = np.array([
               [1, 0.95],
               [0.95, 1]
              ])

# generating independent samples
sample_matrix = norm.rvs(size=(2,sample_size))

# Cholesky decomposition matrix
c = cholesky(cov, lower = True)

# generating correlated samples
data = np.dot(c,sample_matrix)

#plot(data[0], data[1], 'b.')

K = data[0]*30+85

coeff = data[1]*0.2+2

plot(K, coeff, 'b.')
xlabel('Prosumer renewable output')
ylabel('Demand shift coefficient')

df = pd.DataFrame(data)
filepath = 'data.xlsx'
df.to_csv(filepath,index=False)
```
